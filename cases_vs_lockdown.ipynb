```python
%matplotlib inline
import convert_csvs
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
cases = convert_csvs.convert_cases('data/owid-covid-data.csv')
lockdown = convert_csvs.convert_lockdown('data/Global_Mobility_Report.csv')
```

```python
country_code = 'ESP' # Country 3 letters ISO code 
```

```python
fig = plt.figure(figsize=(20, 5))
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
ax = plt.axes()
ax.plot(cases[country_code], color=colors[0], label='Cases p/day')
ax2 = ax.twinx() 
ax2.plot(lockdown[country_code], color=colors[1], label='Stay home')
fig.legend()
```
